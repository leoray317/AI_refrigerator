```python
import pandas as pd
from os.path import isfile, join
from os import listdir

def nan_counter(county_path):
    df_weather = pd.read_excel(county_path)
    #df_weather = df_weather.iloc[:,:8]
    df_weather = df_weather.drop(columns=['縣市','觀測站名'])

    # str date to timestamp and drop the original one
    time_list = pd.to_datetime(df_weather['時間']).tolist()
    df_weather.insert(0, '日期',time_list)
    df_weather = df_weather.drop(columns = ['時間'])

    # transfer str to numeric numbers except some singals
    # coerce - > then invalid parsing will be set as NaN.
    for col in range(1,len(df_weather.columns)):
        df_weather.iloc[:,col] = pd.to_numeric(df_weather.iloc[:,col], errors='coerce')
        print("欄位名稱: ", df_weather.columns.tolist()[col])
        print("缺失數量: ", df_weather.iloc[:,col].isnull().sum())
        print()
        
```

```python


# list all counties' path
combine_data_path = r'../combined_data'
counties = [f for f in listdir(combine_data_path) ]
for county in counties:
    print("=====================================",county[:-5])
    county_path = f'{combine_data_path}/{county}'
    counter = nan_counter(county_path)
    


```
